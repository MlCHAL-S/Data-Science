```python
# testing multiplication by NaN
#df.loc[1, 'horsepower'] = np.nan

df['horsepower*acceleration'] = df.apply(lambda row:
                                         row['horsepower'] 
                                         if pd.isnull(row['acceleration'])
                                         else
                                             row['acceleration']
                                             if pd.isnull(row['horsepower'])
                                             else
                                                 row['horsepower']*row['acceleration'],
                                         axis=1
                                        )
#df = df.drop(['horsepower*acceleration'], axis=1)
df
```

```python
# another way
def row_operations(x):
    if pd.isnull(x['acceleration']):
        return x['horsepower'] 
    else:
        if pd.isnull(x['horsepower']):
            return x['acceleration']
        else:
            return x['horsepower']*x['acceleration']
        
df['horsepower*acceleration'] = df.apply(lambda row: row_operations(row), axis=1)
df
```

```python
# BŁĄD W PANDAS
#bins = pd.IntervalIndex.from_tuples([(df['weight'].min()-1, 2000),(2000, 3600),(3600, df['weight'].max()+1)])
#labels = ['light_weight', 'normal_weight', 'heavy_weight']

#x = pd.cut(df['weight'], bins=bins, labels=labels, ordered=False)

#df['weight_category'] = x

# Funkcja mapująca wartości numeryczne na etykiety słowne
def map_to_label(weight):
    if weight < 2000:
        return 'light_weight'
    elif 2000 <= weight < 3600:
        return 'normal_weight'
    else:
        return 'heavy_weight'

# Przypisz etykiety do kolumny 'weight_category'
df['weight_category'] = df['weight'].apply(lambda x: map_to_label(x))

df
```
